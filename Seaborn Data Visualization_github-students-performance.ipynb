```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5),dpi=200)  

sns.scatterplot(x='math score',y='reading score',data=df,palette='Dark2',hue='gender')
```

```python
plt.figure(figsize=(10,5),dpi=200)  

sns.scatterplot(x='math score',y='reading score',data=df,palette='Dark2',hue='lunch')
```

```python
sns.boxplot(data=df,y='math score')

# we can see distributions of continuous features and their relation to different categories

# BOX PLOT:

# plot a continuous feature against categorical feature(s). If we want sub-categories, use hue option

# continuous features: [math, reading and writing scores]

# Categorical features: [gender, race, parent level of education, lunch, test preparation course]
```

```python
sns.boxplot(data=df,y='math score',x='test preparation course')

# distribution of math score based on whether 'test preparation course' completed/not

# students who have taken 'test preparation course' have high median and marks (quartile values)
```

```python
sns.boxplot(data=df,y='reading score',x='test preparation course')
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education')

# 'math score' distribution as per 'parent education'
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.boxplot(data=df,y='math score',x='parental level of education', 
            hue='test preparation course')

# we can still sub categorize as per the 'test preparation course' by using 'hue'

# students who completed 'test preparation course' have high median, irrespective of 'parents education'

plt.legend(bbox_to_anchor=(1.05,0.95)) # legend placement
plt.show()
```

```python
plt.figure(figsize=(12,6),dpi=200)
sns.boxplot(data=df,x='math score',y='parental level of education', 
            hue='test preparation course')

# changing the orientation of above plot by swapping 'x' and 'y'

# we can still sub categorize as per the 'test preparation course' by using 'hue'

# students who completed 'test preparation course' have high median, irrespective of 'parents education'

plt.legend(bbox_to_anchor=(1.05,0.95)) # legend placement
plt.show()
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='hex')

# default value for kind is 'scatter'

# to see the density of points, two approaches: (i) 
# (i) if kind = 'hex', then as the number of points increases the color shade of the hexagon becomes dark
#              or use kind = 'hist'  this  plots squares in the place of hexagons
# (ii) kind='scatter' with 'alpha' being small
```

```python
sns.jointplot(data=df,x='math score',y='reading score',kind='kde',shade=True)
```

```python
sns.jointplot(data=df,x='math score',y='reading score',hue='gender')

# this is good. two seperate 'kde' for differentcategories
```

```python
# pairplot-- it plots scatter plots for each combination of NUMERICAL columns
#            diagonal wise- histogram of that particular column with itself  

# even if we pass the entire dataframe, pairplot considers only numerical columns and comesup with scatterplots
```

```python
# pairplot- symmetric matrix, hence we can cutoff the duplicate plots above diagonal
```

```python
sns.pairplot(data=df,hue='gender')

# after adding hue (sub-categorization), the diagonal elements are now kde plots (default)
# but if we want histogram itself, we can do it as given in below cell
```

```python
sns.pairplot(data=df,hue='gender',diag_kind='hist')
```
